# 3D Engine Architecture

A three-dimensional (3D) engine is a system used for virtual computer simulations. 3D engines are commonly used in videogames, though other non-entertainment applications also exist. The rendering component of a 3D engine calculates the visual appearance of a scene, while a physics component determines how different objects should interact.
